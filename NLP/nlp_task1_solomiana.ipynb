```python
from nltk.book import text1
import re
from nltk.probability import FreqDist
import matplotlib.pyplot as plt


textN = text1

words = [word.lower() for word in textN if word.isalpha()]

fdist = FreqDist(words)

hapaxes = fdist.hapaxes()
less_than_5 = [word for word, freq in fdist.items() if freq < 5]

print("Hapaxes (unique words):", hapaxes)
print("Words occurring less than 5 times:", less_than_5)

word_lengths = [len(word) for word in fdist.keys()]
word_freqs = list(fdist.values())

plt.figure(figsize=(10, 6))
plt.scatter(word_lengths, word_freqs, alpha=0.5)
plt.title('Mapping from Word Length to Word Frequency')
plt.xlabel('Word Length')
plt.ylabel('Word Frequency')
plt.grid(True)
plt.show()
sentences = ' '.join(textN).split('.')
end_of_sentence_words = [re.findall(r'\b\w+\b', sentence)[-1].lower() for sentence in sentences if sentence.strip() != '' and re.findall(r'\b\w+\b', sentence)]
fdist_end_of_sentence = FreqDist(end_of_sentence_words)
most_common_end_of_sentence = fdist_end_of_sentence.most_common(10)
print("10 most frequent words occurring at the end of the sentence:")
for word, freq in most_common_end_of_sentence:
    print(word, "-", freq)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.text import Text

with open('myk_ist.txt', 'r') as file:
    raw_text = file.read()

tokens = word_tokenize(raw_text)

custom_text = Text(tokens)

print("Concordance for the word 'Миколка':")
custom_text.concordance('Миколка')


print("\nFrequency distribution of words:")
fdist = nltk.FreqDist(custom_text)
print(fdist.most_common(20))
```

```python
import re

def extract_monetary_amounts(text):
    pattern = r'\$\d+\.\d{2}|\$\d+'
    amounts = re.findall(pattern, text)
    return amounts

def convert_to_float(amounts):
    return [float(amount.replace('$', '')) for amount in amounts]

def sum_monetary_amounts(float_amounts):
    return sum(float_amounts)

def main():
    text = "The first amount is $100.45, and the second amount is $349."
    amounts = extract_monetary_amounts(text)

    if amounts:
        float_amounts = convert_to_float(amounts)
        total_amount = sum_monetary_amounts(float_amounts)

        print("Monetary amounts found:", float_amounts)
        print("Total sum of monetary amounts:", total_amount)
    else:
        print("No monetary amounts found in the text.")

if __name__ == "__main__":
    main()
```

```python
import re
import json

json_str = '''
{
    "HTTPHost": "localhost",
    "HTTPPort": 8545,
    "Config": {
        "Enabled": true,
        "Server": "wakuv2.prod",
        "Nodes": ["8.210.222.231", "168.166.125.145"]
    }
}
'''

data = json.loads(json_str)

data['Config']['Nodes'] = [node for node in data['Config']['Nodes'] if not re.match(r'^8\.', node)]

server_name, server_type = re.match(r'^(\w+)\.(\w+)$', data['Config']['Server']).groups()
data['Config']['Server'] = {'name': server_name, 'type': server_type}

json_str_modified = json.dumps(data, indent=4)

print(json_str_modified)
```
